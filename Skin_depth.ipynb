```python
mu_0 = 1.257E-6
N_1 = 1000
N_2 = 100
I_1 = 19E-3
L_1 = 0.3778
B_0 = math.sqrt(2) * mu_0 * N_1 * I_1 / L_1

def b(R):
    res = 2 * math.pi**2 * N_2 * B_0 * R
    return res
```

```python
b(0.4825E-2)
```

```python
def sigma(R,beta):
    res = (b(R)**2) / (beta**2 * mu_0 * math.pi)
    return res
```

```python

print('Brass: Conductivity = {sigma:.2e}'.format(sigma=sigma(0.4975E-2,m2)))
print('Copper: Conductivity = {sigma:.2e}'.format(sigma=sigma(0.5475E-2,m3)))
print('Steel: Conductivity = {sigma:.2e}'.format(sigma=sigma(0.4825E-2,m4)))
```

```python
def delta(sigma, nu):
    res = (math.pi * sigma * mu_0 * nu)**-(1/2)
    return res
```
